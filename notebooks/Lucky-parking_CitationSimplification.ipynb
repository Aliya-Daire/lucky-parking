```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
%matplotlib inline 
```

```python
#load the large size data
df=pd.read_csv("..\data\Parking_Citations_After_July_1_2015.csv",low_memory=False)
df.head()
```

```python
# get 1% of random data from original data
skip = pd.Series(np.random.binomial(1, 0.01, len(df)), index=df.index)
skip
```

```python
#picked the selected data
subset = df[skip==1]
```

```python
#compare the shape of orginal data and simplied data
df.shape, subset.shape
```
