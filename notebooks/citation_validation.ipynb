```python
import numpy as np
import pandas as pd
import datetime as dt

from sodapy import Socrata
```

```python
client = Socrata("data.lacity.org", None)
```

```python
def get_results(date, client):
    results = client.get('wjz9-h9np',
                        where=f'issue_date=="{date}"',
                        limit=50000)
    results_df = pd.DataFrame.from_records(results)
    return results_df
```

```python
past_df = pd.read_csv('../data/citations_ts.csv')
```

```python
date = []
count = []
fines = []

i_date = dt.date(2020, 5, 1)
delta = dt.timedelta(days=1)

while i_date < dt.date.today() - delta:
    # print progression
    if i_date.day == 1:
        print(i_date)
        
    results_df = get_results(i_date, client)
    if len(results_df) > 0:
        if 'fine_amount' in results_df.columns:
            fine_total = np.sum(results_df['fine_amount'].fillna(0).astype(float))  
        else:
            fine_total = 0

        date.append(str(i_date))
        count.append(len(results_df))
        fines.append(fine_total)
    
    i_date += delta
    
df = pd.DataFrame({
    'date' : date,
    'count' : count,
    'fines' : fines
})
```

```python
df_a = past_df[past_df.date_a>='2020-05-01'].reset_index(drop=True)
```

```python
past_df.columns = [c + '_a' for c in past_df.columns]
```

```python
df.columns = [c + '_b' for c in df.columns]
```

```python
df_b = df[df.date_b<='2020-06-26']
```

```python
merged_df = df_a.join(df_b)
```

```python
merged_df['count_diff'] = merged_df['count_b'] - merged_df['count_a']
merged_df['fines_diff'] = merged_df['fines_b'] - merged_df['fines_a']
```
